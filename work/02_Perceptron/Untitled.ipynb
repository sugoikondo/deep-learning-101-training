```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
step(1)
```

```python
# 実現したいこと
def step(x: np.array) -> np.array:
    y = x > 0
    return y.astype(np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
```

```python
y = step(x)
```

```python
plt.ylim(-00.1, 1.1)
```

```python
plt.plot(x, y)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
y = sigmoid(x)

plt.ylim(-0.1, 1.1)
plt.plot(x, y)
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
y = relu(x)

plt.ylim(-0.1, 5)
plt.plot(x, y)
plt.show()
```
